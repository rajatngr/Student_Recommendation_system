```python
# importing required modules
import fitz
import csv
from itertools import dropwhile
from PyPDF2 import PdfWriter, PdfReader
import numpy as np
import pandas as pd

# creating a pdf file object
doc = fitz.open('DKE_handbook.pdf')
```

```python
def pdf_into_list(doc,para):
    ''' funtion to convert pdt into list of subject under each module
        doc: the pdf to be extracted 
        para: delimiter for spliting the text'''
    
    text = ""
    #extracting the text from pdf
    for page in range(13,):
        text+=page.get_text()
        
    # converting the strg into a clean list
    sub_list = []
    lines = text.split(para)
    for line in lines:
        sub_list.append(line)  
    temp_list = []
    for item in sub_list:
        temp_list.append(item.split("\n"))
    
    del temp_list[:2]
    
    
    return text, lines, temp_list
```

```python
def clean_list(final_list):
    '''This function add which lang the course will be 
    held in if it wasnt mentioned
    final_list should be the output list from function pdf_into_list '''
    
    temp_list1 = []

    for items in final_list:
    
        temp_list2 = []
    
        for item in items:
        
            if item.startswith((' M.Sc.','ENG','DEU')):
                temp_list2.append(item)
            
            elif "Team" in item:
                item = 'ANY - '+ item
                temp_list2.append(item)
            
            else: 
                pass
        
        temp_list1.append(temp_list2)

    return temp_list1
```

```python
def header_content(temp_list1):
    '''Separating the items in the list as header and content'''
    header =  []
    for item in temp_list1:
        for i in item:
            a = item[0]
        header.append(a)


    content = []
    for item in temp_list1:
        temp_list = []
        for i in item[1:]:
            temp_list.append(i)
        content.append(temp_list)
        
    return header, content
```

```python
def list_datafrom(header,content):
    '''This fxn converts the list of module and subject into df with rows and column'''
    
    columns = {'header':header,'content':content}
    df = pd.DataFrame(columns)
    df = df.explode("content")
    df[['Degree', 'Module']] = df['header'].str.split('-', 1, expand=True)
    df[['Language', 'Subject']] = df['content'].str.split('-', 1, expand=True)
    df_final = df[['Degree', 'Module','Language','Subject']]
    df_final = df_final.reset_index(drop=True)
    return df_final
```

```python
text, lines, final_list = pdf_into_list(doc,'FIN:')
clean_list = clean_list(final_list)
header, content = header_content(clean_list)
Module_df = list_datafrom(header,content)
```
